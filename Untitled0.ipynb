```python
import pandas
```

```python
import matplotlib.pyplot as plt
from random import randrange, choice
import math
```

```python
total_supply = 10000000000
fee_rate = 0.001
ether_bal = 0
in_circ = 0

balances = [0,0,0]

column_names = ["total_supply", "ether_bal", "in_circ", "exchange_rate", "account_1", "account_2", "account_3"]
df = pandas.DataFrame(columns = column_names, data = [[total_supply, ether_bal, in_circ, 1, balances[0], balances[1], balances[2]]])

def buy_coyne_coin(ether_amt, account): 
  global fee_rate
  global total_supply
  global in_circ
  global ether_bal
  global balances
  global df
  
  exchange_rate = getExchangeRate()

  coyne_coin = ether_amt / exchange_rate
  fee = coyne_coin * fee_rate

  total_supply = total_supply - coyne_coin

  in_circ = in_circ + coyne_coin

  ether_bal = ether_bal + ether_amt 

  balances[account] = balances[account] + (coyne_coin - fee)
  df.loc[len(df.index)] = [total_supply, ether_bal, in_circ, exchange_rate, balances[0], balances[1], balances[2]]



def transfer_coyne_coin(coyne_coin_amt, to_account, from_account):
  global fee_rate
  global balances
  global total_supply
  global df

  fee = coyne_coin_amt * fee_rate
  recipient_amt = coyne_coin_amt - fee
  total_supply = total_supply + fee

  balances[from_account] = balances[from_account] - coyne_coin_amt
  balances[to_account] = balances[to_account] + recipient_amt
  df.loc[len(df.index)] = [total_supply, ether_bal, in_circ, getExchangeRate(), balances[0], balances[1], balances[2]]


def redeem(coyne_coin_amt, account):
  global ether_bal
  global total_supply
  global fee_rate
  global in_circ
  global balances
  global df

  total_supply = total_supply + coyne_coin_amt

  exchange_rate = getExchangeRate()
  redeem_amount = coyne_coin_amt / exchange_rate

  fee = redeem_amount * fee_rate
  return_amount = redeem_amount - fee

  ether_bal = ether_bal - return_amount

  in_circ = in_circ - coyne_coin_amt

  balances[account] = balances[account] - coyne_coin_amt
  df.loc[len(df.index)] = [total_supply, ether_bal, in_circ, exchange_rate, balances[0], balances[1], balances[2]]

def getExchangeRate(): 
  global in_circ 
  global ether_bal
  if(ether_bal == 0):
    return 1000/1001
  exchange_rate = in_circ / ether_bal # total_supply or in_circ?
  return exchange_rate

def choose_transaction_type(account):
  global balances
  if(balances[account] == 0):
    return 0
  return randrange(3)

def choose_account():
  return randrange(3)

def choose_amount(account, transaction_type):
  # buy
  global total_supply
  global balances
  if(transaction_type == 0):
    return randrange(math.floor(total_supply))
  # transfer
  if(transaction_type == 1):
    return randrange(math.floor(balances[account]))
  # redeem
  if(transaction_type == 2):
    return randrange(math.floor(balances[account]))


def choose_other_account(account):
  possible = [0,1,2]
  possible.remove(account)
  other_account = choice(possible)
  return other_account


for i in range(0,1000): 
  # transaction_type = 0 - buy, 1 - transfer, 2 - redeem
  account = choose_account()
  transaction_type = choose_transaction_type(account)
  amount = choose_amount(account, transaction_type)
  if(transaction_type == 0):
    buy_coyne_coin(amount, account)
  if(transaction_type == 1):
    other_account = choose_other_account(account)
    transfer_coyne_coin(amount, other_account, account)
  if(transaction_type == 2):
    redeem(amount, account)

  # for a million iterations
    # randomly choose a transaction type
    # randomly choose an account(s)
    # randomly amount for transaction type (with limitations based on balances)
    # execute the transaction 
    # record vars in dataframe
```
